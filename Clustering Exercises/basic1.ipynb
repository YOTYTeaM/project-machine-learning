```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Source: https://www.kaggle.com/datasets/joonasyoon/clustering-exercises?select=basic1.csv

```python
dataset = pd.read_csv("basic1.csv")
dataset.head()
```

```python
plt.scatter(dataset["x"], dataset["y"])
plt.show()
```

```python
dataset['color'].value_counts()
```

```python
dataset.dropna(inplace=True)
```

```python
x = dataset[['x', 'y']].values
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

scaler = MinMaxScaler()
model = KMeans(n_clusters=4)
pipeline = Pipeline([('scaler', scaler), ('model', model)])
pipeline.fit_predict(x)

print(f"score {silhouette_score(x, pipeline.steps[1][1].labels_, metric='euclidean')}")
```

```python
labels = pipeline.steps[1][1].labels_
plt.scatter(x[:,0], x[:, 1], c=labels)
plt.show()

plt.scatter(dataset["x"], dataset["y"], c=dataset["color"])
plt.show()
```
